```python
import cv2  # OpenCV for image processing
import pytesseract  # OCR engine
from PIL import Image  # Image handling
import numpy as np

# Set the path for Tesseract-OCR (Windows users)
# Uncomment the below line and provide the correct path if needed.
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
# Function to preprocess image
def preprocess_image(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Convert to grayscale (helps OCR accuracy)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to remove noise
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    # Save the processed image (optional)
    processed_path = "processed_image.png"
    cv2.imwrite(processed_path, thresh)

    return processed_path

# Function to extract text using OCR
def extract_text(image_path):
    # Preprocess the image
    processed_path = preprocess_image(image_path)

    # Open the processed image using PIL
    img = Image.open(processed_path)

    # Apply OCR to extract text
    extracted_text = pytesseract.image_to_string(img)

    return extracted_text

# Test the OCR function
image_path =r"C:\Users\annad\Documents\Automated-Paper-Evaluation-system\Biology-0089-0159\0098.jpg"  # Change this to your image file path
text = extract_text(image_path)
if image is None:
    raise FileNotFoundError(f"Error: Image not found at {image_path}")


# Print extracted text
print("Extracted Text:\n", text)
```
